```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
```

```python
# loading the data
```

```python
data = pd.read_csv('i2w_train.csv')
data
```

```python
# tcol -> lat, trow -> long

x, y = data['trow']/1280.0, data['lon']/3-118.0
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.25)
```

```python
# creating a model
model = keras.Sequential(
    [
        keras.layers.Dense(1, input_shape=[1], activation="relu", name="output"),
        
    ]
)
model.summary()
```

```python
# batch_size = 128
epochs = 10000

model.compile(loss="mse", optimizer=tf.keras.optimizers.SGD(lr=1e-3, momentum=0.9))
model.fit(xtrain, ytrain, batch_size=None, epochs=epochs, validation_split=0.1)
```
