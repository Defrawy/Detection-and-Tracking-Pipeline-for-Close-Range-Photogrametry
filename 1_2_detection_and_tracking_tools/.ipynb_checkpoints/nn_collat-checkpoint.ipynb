```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
```

```python
# loading the data
```

```python
data = pd.read_csv('i2w_train.csv')
data
```

```python
data.corr()
```

```python
# tcol -> lat, trow -> long
from sklearn.preprocessing import MinMaxScaler

print(data[['trow', 'tcol', 'lat']])

data = MinMaxScaler().fit_transform(data[['trow', 'tcol', 'lat']])
print(data)
x, y = data[:, [0, 1]], data[:,-1]
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.30)
```

```python
# creating a model
model = keras.Sequential(
    [
        keras.layers.Dense(10, input_shape=[2], activation="relu"),
        keras.layers.Dense(10, activation="relu"),
        keras.layers.Dense(1, name="output")
        
    ]
)
model.summary()
```

```python
# batch_size = 128
epochs = 100000

model.compile(loss="mse", optimizer=tf.keras.optimizers.SGD(lr=1e-5, momentum=0.9))
H = model.fit(xtrain, ytrain, batch_size=None, epochs=epochs, validation_split=0.1, verbose=1)
```

```python
model.weights
```

```python
evl = model.evaluate(xtest, ytest)
print(evl)
```

```python
model.save('nn_collat10_10_1.h5')
```
