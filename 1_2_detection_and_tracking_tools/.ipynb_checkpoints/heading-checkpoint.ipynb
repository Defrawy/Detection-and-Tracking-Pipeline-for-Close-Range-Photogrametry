```python
track = track[track['id'] == boat]
track['x'] = track['x'] + track['w']
track['y'] = track['y'] + track['h'] - 420
track
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(10, 4))
plt.scatter( track['x'], -1 * track['y'])
plt.title('Video 9 Boat Trajector Pixel Wise')
plt.xlabel('X Pixels')
plt.ylabel('Y Pixels')
```
