```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import load_model
```

```python
lat = load_model('nn_collat10_10_1.h5')
lon = load_model('nn_rowlon10_10_1.h5')
```

```python
data = pd.read_csv('i2w_pred.csv')
```

```python
from sklearn.preprocessing import MinMaxScaler

print(data[['plon', 'lon']])
scaler = MinMaxScaler().fit(data[['tcol', 'trow', 'lat','lon']])
data = scaler.transform(data[['plat', 'plon', 'lon']])
print(data)
x, y = data[:, [0, 1]], data[:, -1]
# xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.25)
```

```python
plon = scaler.inverse_transform(np.concatenate([lon.predict(x), lon.predict(x), lon.predict(x)], axis=1))
plon = plon[:, -1]
plon
```

```python
data = pd.read_csv('i2w_pred.csv')
data['pplon'] = plon
data.corr()
data.to_csv('predictions6.csv')
```
