```python
from io import BytesIO
import requests
import pandas as pd
from arcgis.features import FeatureLayer
from arcgis.gis import GIS
from arcgis.mapping import WebMap
```

```python
gis = GIS("home")
```

```python
items = gis.content.search("Coronavirus_2019_nCoV_Cases owner:CSSE_GISandData", outside_org=True)
print(items)
```

```python
items[0]
```

```python
src_url = "https://services1.arcgis.com/0MSEUqKaxRlEPj5g/arcgis/rest/services/Coronavirus_2019_nCoV_Cases/FeatureServer/1"
fl = FeatureLayer(url=src_url)
```

```python
df_global = fl.query(where="1=1",
                     return_geometry=True,
                     as_df=True)
```

```python
df_global.head()
```

```python
df_global[~pd.isnull(df_global['Province_State'])].groupby('Country_Region').sum()[['Confirmed', 'Recovered', 'Deaths']]
```

```python
fset_usa = fl.query(where="Country_Region='US'")
fset_usa_df = fl.query(where="Country_Region='US'", as_df = True)
fset_usa_df.head()
```

```python
src_url = 'https://services.arcgis.com/P3ePLMYs2RVChkJx/arcgis/rest/services/USA_States_Generalized/FeatureServer/0'
fl_usa = FeatureLayer(url=src_url)
us_states_df = fl_usa.query(where="1=1",
                     return_geometry=True,
                     as_df=True)
us_states_df.head()
```

```python
df_usa = fset_usa.sdf[['Province_State', 'Confirmed', 'Recovered', 'Deaths']]
df_usa = df_usa.assign(STATE_NAME = df_usa["Province_State"])
df_usa.head()
```

```python
cols_4 = ['STATE_NAME','SHAPE']
overlap_rows_usa = pd.merge(left = us_states_df[cols_4], right = df_usa, 
                            how='inner', on = 'STATE_NAME')
overlap_rows_usa.head()
```

```python
map_usa = gis.map("US", zoomlevel=4)
map_usa
```

```python
map_usa.add_layer(fl, { "type": "FeatureLayer",
                     "renderer":"ClassedSizeRenderer",
                     "field_name":"Confirmed"})

```

```python
map_usa.legend = True
```

```python
map_usa_poly = gis.map("US")
map_usa_poly.clear_graphics()
overlap_rows_usa.spatial.plot( kind='map', map_widget=map_usa_poly,
                               renderer_type='c',  # for class breaks renderer
                               class_count=4,  # choose the number of classes
                               method='esriClassifyEqualInterval',  # classification algorithm
                               col='Confirmed',  # numeric column to classify
                               cmap='inferno',  # color map to pick colors from for each class
                               alpha=0.7  # specify opacity
                              )

map_usa_poly
```

```python
map_usa_poly.zoom = 4
map_usa_poly.legend = True
```
